```python
from typing import TypedDict, Annotated, Optional, List
from langgraph.graph import add_messages, StateGraph, END
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv
from langchain_community.tools import TavilySearchResults
from langgraph.checkpoint.memory import MemorySaver
from uuid import uuid4
from langchain_core.messages import AIMessage, HumanMessage, ToolMessage
from IPython.display import Image, display
from langchain_core.runnables.graph import MermaidDrawMethod
```

```python
uuid4().int
```

```python
load_dotenv()
```

```python
llmGemini=ChatGoogleGenerativeAI(model="gemini-2.0-flash-001")
llmOpenAI=ChatOpenAI()
```

```python
search_tool=TavilySearchResults(max_results=3)
```

```python
tools=[search_tool]
```

```python
memory=MemorySaver()
```

```python
model_with_tools=llmGemini.bind_tools(tools=tools)
```

```python
model_with_tools.invoke(input="Write a short poem for my 10 month old son")
```

```python
model_with_tools.invoke(input="Tell me more about Operation Sindoor")
```

```python
search_tool.invoke(
    input=model_with_tools.invoke(input="Tell me more about Operation Sindoor").tool_calls[0]['args']
)
```

```python
search_tool.invoke(
    input=model_with_tools.invoke(input="Tell me more about Operation Sindoor").tool_calls[0]['args']['query']
)
```

```python
class State(TypedDict):
    messages: Annotated[List,add_messages]
```

```python
async def model(state:State):
    result=await model_with_tools.ainvoke(input=state["messages"])
    state['messages']=[result]
    return state
```

```python
async def tools_router(state: State):
    last_message=state["messages"][-1]
    return "tool_node" if hasattr(last_message,"tool_calls") and len(last_message.tool_calls)>0 else "end"
```

```python
async def tool_node(state):
    """
        Custom tool node that handles tool calls from the LLM
    """
    tool_calls=state["messages"][-1].tool_calls
    # print(f"Last Message: {state['messages'][-1]}")
    
    # Initialize list to store tool messages
    tool_messages=[]

    # Process each tool call
    for tool_call in tool_calls:
        tool_name=tool_call["name"]
        tool_query=tool_call["args"]['query']
        tool_id=tool_call["id"]

        # Handle the search tool
        if tool_name=="tavily_search_results_json":
            # Execute the search tool with the provided arguments
            search_results=await search_tool.ainvoke(input=tool_query)

            # Create a ToolMessage for this result
            tool_message=ToolMessage(
                content=str(search_results),
                tool_call_id=tool_id,
                name=tool_name
            )
            tool_messages.append(tool_message)
    # print(tool_messages)
    state['messages']=tool_messages

    return state
```

```python
graph_builder=StateGraph(state_schema=State)
```

```python
graph_builder.add_node(node="model",action=model)
```

```python
graph_builder.add_node(node="tool_node",action=tool_node)
```

```python
graph_builder.add_edge(start_key="tool_node",end_key="model")
```

```python
graph_builder.set_entry_point(key="model")
```

```python
graph_builder.add_conditional_edges("model",path=tools_router,
                                    path_map={"tool_node":"tool_node","end":END}
                                   )
```

```python
graph=graph_builder.compile(checkpointer=memory)
```

```python
display(Image(data=graph.get_graph().draw_mermaid_png(draw_method=MermaidDrawMethod.API
)))
```
